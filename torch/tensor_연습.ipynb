print(b)의 결과가 다음과 같이 되는 code를 작성하시오.  
b는 a, a1, a2 를 이용하여 정의함.   
 , 유무는 신경 쓰지 말것

1. [[2., 2., 2., 5.],  
        [5., 5., 2., 5.],  
        [9., 5., 5., 2.]]  

```python
import torch
a = torch.tensor([[[ 2., 2., 2., 5.]],
                  [[ 5., 5., 2., 5.]],
                  [[ 9., 5., 5., 2.]]])
b = a.squeeze()

print(b)
```

2. [[[1., 2., 2.]],  
    [[3., 4., 4.]]]

```python
import torch
a = torch.FloatTensor([[1, 2, 2],
                       [3, 4, 4]])

b = a.unsqueeze(0)
print(b)
print(b.size())
```

3. [[ 1.,  3., 10., 12.],  
      [ 4.,  6., 13., 15.],  
      [ 7.,  9., 16., 18.]]  

```python
import torch
a1 = torch.FloatTensor([[1, 3],
                        [4, 6],
                        [7, 9]])

a2 = torch.FloatTensor([[10, 12],
                        [13, 15],
                        [16, 18]])

b = torch.cat([a1, a2], dim = 1)

print(b)
print(b.size())
```
