```python
import matplotlib.pyplot as plt
import tensorflow as tf

(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.mnist.load_data()

print(train_images.shape) # 28 x 28 픽셀 6만 장
print(train_labels)       # 각 이미지의 정답

print(test_images.shape)  # 28 x 28 픽셀 1만 장
print(test_labels)        # 각 이미지의 정답
```

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(512, activation='sigmoid'))

# softmax
# 다중 클래스 분류 문제에서 각 클래스에 대한 확률을 출력하는 활성화 함수
# 총합이 1이 되도록 각 클래스에 대한 확률을 계산하여 출력
model.add(tf.keras.layers.Dense(10, activation='softmax'))

model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
train_images = train_images.reshape((60000, 784))       # 2차원 이미지를 1차원 벡터로 변환
train_images = train_images.astype('float32') / 255.0   # 픽셀 값을 0과 1 사이로 정규화

test_images = test_images.reshape((10000, 784))
test_images = test_images.astype('float32') / 255.0
```

```python
train_labels = tf.keras.utils.to_categorical(train_labels) # 정수 레이블을 원핫 인코딩으로 변환
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
model.fit(train_images, train_labels, epochs=5, batch_size=256)

test_loss, test_acc = model.evaluate(test_images, test_labels)
print('테스트 정확도:', test_acc)

plt.show()
```
