## Arithmetic Operations
---

```python
import numpy as np

a = np.arange(1 ,11)
b = np.arange(10)
```

```python
print(a)
```

```python
print(b)
```

---
Subtraction

```python
print(a - b)
```

```python
print(np.subtract(a, b))
```

---
Addition

```python
print(a + b)
```

```python
print(np.add(a, b))
```

---
Division

```python
print(a / b)
```

```python
print(np.divide(a, b))
```

---
Multiplication

```python
print(a * b)
```

```python
print(np.multiply(a, b))
```
