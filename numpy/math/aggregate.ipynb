## Aggregate Functions
---

```python
import numpy as np

a = np.arange(10)
b = np.arange(10)
```

```python
a.sum()
np.sum(a)
```

```python
a.min()
np.min(a)
```

```python
a.max()
np.max(a)
```

```python
a.cumsum()
```

```python
a.mean()
np.mean(a)
```

```python
np.median(a)
```

```python
## 상관계수 계산
np.corrcoef(a, b)
```

```python
a.std()
np.std(a)
```
