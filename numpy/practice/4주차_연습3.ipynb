5. 다음 code의 배열 a에서 내부 영역만 slicing한 뒤,  
그 안에서 전체 평균보다 큰 값만 남기고,  
그 값들 중 최솟값은 제외한 나머지의 평균을 구하라.  



```python
import numpy as np

a = np.arange(1, 26).reshape(5, 5)

b = a[1:4, 1:4]
b = b[b > b.mean()]
b = b[b != b.min()]
print(b.mean())
```

6. 다음 code에 정의된  a를 이용하여  아래 결과 배열 b를 만들고 print하시오.
